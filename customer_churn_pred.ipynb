```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train=pd.read_csv(r'/content/drive/MyDrive/Colab_Notebooks/dataset/train (7).csv')
test=pd.read_csv(r'/content/drive/MyDrive/Colab_Notebooks/dataset/test (5).csv')
test.head()
train.head()
```

```python
train['churn'] = train['churn'].map({'yes':1,'no':0})
```

```python
train.info()
```

```python
train['churn'].value_counts()
```

```python
test['churn']=5
```

```python
test.drop('id',axis=1,inplace=True)
```

```python
df = pd.concat([train, test],axis=0)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['area_code'].value_counts()
```

```python
df.drop('area_code',axis=1,inplace=True)
```

```python
cat=df.select_dtypes(include='object')
cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat.columns:
    cat[i]=le.fit_transform(cat[i])
cat.head()
```

```python
cat_col=df.select_dtypes(include='object')
df=df.drop(cat_col.columns,axis=1)
df.head()
```

```python
df=pd.concat([df,cat],axis=1)
df.head()
```

```python
test = df[df['churn']==5]
train = df[df['churn']!=5]
```

```python
x = train.drop('churn',axis=1)
y = train['churn']
```

```python
from sklearn.preprocessing import StandardScaler
xcol=x.columns
sc=StandardScaler()
x=sc.fit_transform(x)
x=pd.DataFrame(x,columns=xcol)
x.head()
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x,y=smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
prediction1=dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print (classification_report(y_test,prediction1))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
prediction2=rf.predict(x_test)
```

```python

print (classification_report(y_test,prediction2))
```

```python
rf.score(x_train,y_train),rf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
gb.score(x_train,y_train),gb.score(x_test,y_test)
```
